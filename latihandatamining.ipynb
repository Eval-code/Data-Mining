# Pertemuan 3 – Data Preprocessing

```python

import pandas as pd
import numpy as np

# Contoh dataset
data = {'Umur': [25, 30, np.nan, 40, 35],
        'Pendapatan': [5000, 7000, 8000, None, 6000],
        'Gender': ['L', 'P', 'P', 'L', None]}

df = pd.DataFrame(data)
print("Data awal:\n", df)

# Isi missing value dengan mean/median
df['Umur'].fillna(df['Umur'].mean(), inplace=True)
df['Pendapatan'].fillna(df['Pendapatan'].median(), inplace=True)

# Encoding Gender
df['Gender'] = df['Gender'].map({'L': 0, 'P': 1})

print("\nData setelah preprocessing:\n", df)
```

# Pertemuan 4 – Exploratory Data Analysis (EDA)

```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df = iris.frame

print(df.describe())

# Visualisasi distribusi
sns.histplot(df['sepal length (cm)'], kde=True)
plt.show()

# Scatter plot
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df)
plt.show()
```

# Pertemuan 5 – Klasifikasi (Decision Tree)

```python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Akurasi:", accuracy_score(y_test, y_pred))
```

# Pertemuan 6 – Klasifikasi (KNN & SVM)

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("Akurasi KNN:", accuracy_score(y_test, knn.predict(X_test)))

# SVM
svm = SVC()
svm.fit(X_train, y_train)
print("Akurasi SVM:", accuracy_score(y_test, svm.predict(X_test)))
```

# Pertemuan 7 – Clustering (K-Means)

```python

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

import pandas as pd
df = pd.DataFrame(X, columns=iris.feature_names)
df['Cluster'] = clusters

print(df.head())

sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['Cluster'], palette="deep")
plt.show()
```

# Pertemuan 9 – Asosiasi (Apriori)

```python

from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

dataset = [
    ['roti', 'susu'],
    ['roti', 'telur', 'susu'],
    ['roti', 'kopi'],
    ['telur', 'kopi']
]

# One-hot encoding
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apriori
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

print(rules)
```

# Pertemuan 10 – Regresi (Linear)

```python

from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

data = fetch_california_housing(as_frame=True)
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

reg = LinearRegression()
reg.fit(X_train, y_train)

print("Score R^2:", reg.score(X_test, y_test))
```

# Pertemuan 11 – Text Mining (TF-IDF + Sentiment)

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

docs = [
    "Film ini sangat bagus dan menyenangkan",
    "Film ini jelek sekali dan membosankan",
    "Sangat seru dan menyentuh",
    "Tidak bagus, mengecewakan"
]
labels = [1, 0, 1, 0]  # 1 = positif, 0 = negatif

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

model = MultinomialNB()
model.fit(X, labels)

print("Prediksi:", model.predict(vectorizer.transform(["Film ini keren banget!"])))
```
